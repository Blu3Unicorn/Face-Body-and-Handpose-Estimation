```python
import mediapipe as mp
import cv2
import numpy as np
import uuid
import os
```

```python
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands
mp_drawing_styles = mp.solutions.drawing_styles
mp_face_mesh = mp.solutions.face_mesh
mp_pose = mp.solutions.pose
```

```python
cap = cv2.VideoCapture(0)

hands =  mp_hands.Hands(max_num_hands = 10, min_detection_confidence=0.8, min_tracking_confidence=0.5)
face_mesh = mp_face_mesh.FaceMesh(static_image_mode=True, max_num_faces=5, refine_landmarks=True, min_detection_confidence=0.5)
pose = mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5)

while cap.isOpened():

            #frame represents image from webcam
    ret, frame = cap.read()

            #transforming color to RGB
            #detection model needs  RGB
    image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

            #writeable flag is now false
    image.flags.writeable = False

            #detections
            #we pass the image to hands and makes the detections
            #array of info about the position of each hand landmark
    results = hands.process(image)
    results2 = face_mesh.process(image)
    results3 = pose.process(image)


            #set flag to true again
    image.flags.writeable = True

            #convert back to BGR
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

            #print detection results
    print(results)


    if results.multi_hand_landmarks:
        for num, hand in enumerate(results.multi_hand_landmarks):
            mp_drawing.draw_landmarks(image, hand, mp_hands.HAND_CONNECTIONS)
            
    if results2.multi_face_landmarks:
        for face_landmarks in results2.multi_face_landmarks:
            mp_drawing.draw_landmarks(
                image=image,
                landmark_list=face_landmarks,
                connections=mp_face_mesh.FACEMESH_TESSELATION,
                landmark_drawing_spec=None,
                connection_drawing_spec=mp_drawing_styles
                .get_default_face_mesh_tesselation_style())
            mp_drawing.draw_landmarks(
                image=image,
                landmark_list=face_landmarks,
                connections=mp_face_mesh.FACEMESH_CONTOURS,
                landmark_drawing_spec=None,
                connection_drawing_spec=mp_drawing_styles
                .get_default_face_mesh_contours_style())
            mp_drawing.draw_landmarks(
                image=image,
                landmark_list=face_landmarks,
                connections=mp_face_mesh.FACEMESH_IRISES,
                landmark_drawing_spec=None,
                connection_drawing_spec=mp_drawing_styles
                .get_default_face_mesh_iris_connections_style())

    mp_drawing.draw_landmarks(
        image,
        results3.pose_landmarks,
        mp_pose.POSE_CONNECTIONS,
        landmark_drawing_spec=mp_drawing_styles.get_default_pose_landmarks_style())


            #renders the image from webcam
    cv2.imshow('Tracking', image)

            #if user presses q, then the webcam feed stops
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
